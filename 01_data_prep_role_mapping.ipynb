```python
import pandas as pd
import numpy as np

# 1. Load your existing modern dataset
# Replace with your actual file path
df = pd.read_csv('data-1.csv')

# 2. Keep only the necessary features to keep the dataset manageable
# We keep identifiers, physical traits, and key technical stats to find "what made them GOATs"
features_to_keep = [
    'Name', 'Age', 'Nationality', 'Overall', 'Position', 
    'Finishing', 'ShortPassing', 'Dribbling', 'BallControl', 
    'SprintSpeed', 'Stamina', 'StandingTackle', 'Interceptions', 'Vision'
]
df_clean = df[features_to_keep].copy()

# 3. Create historical legends (1970s to 1990s) to satisfy the era constraint
# Stats are estimated to FIFA scales for these historical GOATs
historical_players = pd.DataFrame({
    'Name': ['Pelé', 'Johan Cruyff', 'Franz Beckenbauer', 'Diego Maradona', 'Paolo Maldini'],
    'Age': [29, 27, 28, 26, 28], # Prime ages
    'Nationality': ['Brazil', 'Netherlands', 'Germany', 'Argentina', 'Italy'],
    'Overall': [98, 94, 93, 97, 94],
    'Position': ['ST', 'CF', 'CB', 'CAM', 'LB'],
    'Finishing': [97, 88, 70, 91, 55],
    'ShortPassing': [89, 93, 90, 95, 85],
    'Dribbling': [96, 95, 82, 97, 70],
    'BallControl': [95, 94, 88, 98, 82],
    'SprintSpeed': [93, 89, 80, 90, 85],
    'Stamina': [88, 86, 92, 85, 90],
    'StandingTackle': [45, 60, 95, 40, 96],
    'Interceptions': [50, 65, 96, 45, 95],
    'Vision': [93, 96, 90, 97, 75]
})

# Combine modern players with historical legends
df_combined = pd.concat([df_clean, historical_players], ignore_index=True)

# 4. Map traditional football positions to BAGA Roles
def map_baga_role(position):
    attackers = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW']
    # Grouping playmakers and midfielders as supporters
    supporters = ['LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB']
    # Grouping defenders and Goalkeeper
    defenders = ['LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']
    
    if position in attackers:
        return 'Attacker'
    elif position in defenders:
        return 'Defender'
    elif position in supporters:
        return 'Supporter'
    else:
        return 'Unknown'

df_combined['BAGA_Role'] = df_combined['Position'].apply(map_baga_role)

# 5. Export to Excel (Mandatory for the assignment)
# You will attach this file to your final submission report
df_combined.to_excel('1_Data_Prep_and_Roles.xlsx', index=False)
print("Data preparation complete. Exported to 1_Data_Prep_and_Roles.xlsx")
```
